```python
df['season'] = df['season'].astype(str)
```

```python
def fix_season(x):
    if x == '2007/08':
        return 2008
    elif x == '2009/10':
        return 2010
    elif x == '2020/21':
        return 2020
    else:
        return int(x)

df['season'] = df['season'].apply(fix_season)
```

```python
def clean_venue(x):

    x = str(x)

    if 'Eden Gardens' in x:
        return 'Eden Gardens'

    elif 'Wankhede' in x:
        return 'Wankhede Stadium'

    elif 'Chinnaswamy' in x:
        return 'M Chinnaswamy Stadium'

    elif 'Chidambaram' in x:
        return 'MA Chidambaram Stadium'

    elif 'Rajiv Gandhi International' in x:
        return 'Rajiv Gandhi International Stadium'

    elif 'Arun Jaitley' in x or 'Feroz Shah Kotla' in x:
        return 'Arun Jaitley Stadium'

    elif 'Sawai Mansingh' in x:
        return 'Sawai Mansingh Stadium'

    elif 'Punjab Cricket Association' in x:
        return 'Punjab Cricket Association Stadium'

    elif 'Dr DY Patil' in x or 'DY Patil' in x:
        return 'Dr DY Patil Sports Academy'

    elif 'Maharashtra Cricket Association' in x:
        return 'Maharashtra Cricket Association Stadium'

    elif 'Dr. Y.S. Rajasekhara Reddy' in x:
        return 'ACA-VDCA Cricket Stadium'

    elif 'Brabourne' in x:
        return 'Brabourne Stadium'

    elif 'Narendra Modi' in x or 'Sardar Patel' in x:
        return 'Narendra Modi Stadium'

    elif 'Ekana' in x or 'Bharat Ratna' in x:
        return 'Ekana Cricket Stadium'

    elif 'Maharaja Yadavindra' in x:
        return 'Maharaja Yadavindra Singh International Cricket Stadium'

    elif 'Himachal Pradesh Cricket Association' in x or 'Dharamsala' in x:
        return 'Himachal Pradesh Cricket Association Stadium'

    elif 'Sheikh Zayed' in x or 'Zayed Cricket Stadium' in x:
        return 'Sheikh Zayed Stadium'

    else:
        return x
```

```python
df['venue'] = df['venue'].apply(clean_venue)
```
